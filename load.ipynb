```python
import torch
from torch import nn
from vis import LSTMModel
from start import StockDataManager
```

```python
CHECKPOINT_PATH = 'stock_predictor_model.pth'
checkpoint = torch.load(CHECKPOINT_PATH, weights_only=False)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
data_new = StockDataManager().prepare_for_training(symbols=['AAPL'], seq_length=60, prediction_days=1, total_days=365, train_ratio=0.8, normalize=True, save=False)
```

```python
model = LSTMModel()
model.to(device)
model.create(data_new['X_train'], checkpoint['struct'])
model.load_state_dict(checkpoint['model_state_dict'])
model.eval()
```
